```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sklearn.linear_model import LinearRegression
```

```python
os.chdir('/home/lenovo/data files')
```

```python
df = pd.read_csv('baseball.csv')
```

```python
plt.scatter(df['W'],df['Playoffs'])
plt.xlabel('wins')
plt.ylabel('playoffs')
```

```python
difference=np.subtract(df['RS'],df['RA'])
```

```python
#scatter plot between run-difference and wins

plt.scatter(difference,df['W'])
plt.xlabel('run-difference')
plt.ylabel('wins')
```

```python
mean_wins=np.mean(df['W'])
```

```python
z_difference=stats.zscore(difference)
```

```python
z_wins=stats.zscore(df['W'])
```

```python
#standarized scatter plot between run-difference and wins

plt.scatter(z_difference,z_wins)
plt.xlabel('run-difference')
plt.ylabel('wins')
```

```python
normalise_difference=np.divide((np.subtract(z_difference,np.min(z_difference))),(np.subtract(np.max(z_difference),np.min(z_difference))))
```

```python
normalise_wins=np.divide((np.subtract(z_wins,np.min(z_wins))),(np.subtract(np.max(z_wins),np.min(z_wins))))
```

```python
#normalized scatter plot between run-difference and wins

plt.scatter(normalise_difference,normalise_wins)
plt.xlabel('run-difference')
plt.ylabel('wins')
```

```python
x_train=normalise_difference.reshape(-1,1)
y_train=normalise_wins.reshape(-1,1)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```
